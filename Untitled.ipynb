```python
power_b1 = hsc3x2pt.power_b1_class()
```

```python
t0 = time()
k = np.logspace(-4, 1, 100)
pklin = power_b1.get_Pmm(k, 0.1, model='lin')
pkhalo = power_b1.get_Pmm(k, 0.1, model='nonlin')
t1 = time()

print(f't = {t1-t0} sec')

plt.figure()
plt.loglog(k, pklin)
plt.loglog(k, pkhalo)
plt.show()
```

```python
z_min, z_max = 1e-4, 1.1
l = 2

t0 = time()
z, chi, k, p_lin  = power_b1.get_Pmm_lz(l, z_min, z_max, model='lin')
t1= time()
z, chi, k, p_nonlin = power_b1.get_Pmm_lz(l, z_min, z_max, model='nonlin')
t2 = time()
print(f't = {t1-t0} sec')
print(f't = {t2-t1} sec')

plt.figure()
plt.xlabel(r'$k=l/\chi$')
plt.ylabel(r'$P(l/\chi, z(\chi))$')
plt.loglog(k, p_lin)
plt.loglog(k, p_nonlin)
plt.show()
```

```python
convolve2D = hsc3x2pt.convolve2D_class(power_b1)
```

```python
w_s = hsc3x2pt.galaxy_sample_source_class(['s1', 100, 0.01, 1.0])
convolve2D.set_galaxy_sample(w_s)
w_g = hsc3x2pt.galaxy_sample_lens_class(['l1', 55, 50, 60, 2.1, 1.0, 3.0])
convolve2D.set_galaxy_sample(w_g)
w_g = hsc3x2pt.galaxy_sample_lens_class(['l2', 50, 40, 55, 2.1, 1.0, 3.0])
convolve2D.set_galaxy_sample(w_g)

convolve2D.set_cosmology_from_dict()
```

```python
%%time
l = np.logspace(-2, 3, 100)
convolve2D.compute_all_P2D(l, progress=True)
```

```python
plt.figure()
plt.xlabel(r'$l$')
plt.ylabel(r'$l(l+1)C(l)$')
for key, P2D in convolve2D.P2Dcache.items():
    #plt.loglog(P2D['l'], P2D['P2D']/max(P2D['P2D']) * l*(l+1), label=key)
    if max(P2D['P2D'])!=0:
        plt.loglog(P2D['l'], P2D['P2D'], label=key)
plt.legend()
plt.show()
```
